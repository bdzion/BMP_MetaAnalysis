```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the box plots
# Tile flow as percentage of annual precipitation
field_scale_precipitation = [16, 41, 26, 66, 11, 40, 17, 22, 13.7, 21.9, 17.5, 32]
watershed_scale_precipitation = [8.5, 16.2, 13, 19, 34.14, 45]

# Tile flow as percentage of total flow
field_scale_total_flow = [78, 90, 80, 97, 81, 96, 83, 90]
watershed_scale_total_flow = [42, 53, 80, 40, 45, 47]

# Combine the data for the single plot
data = [
    field_scale_precipitation, 
    watershed_scale_precipitation, 
    field_scale_total_flow, 
    watershed_scale_total_flow
]

# Number of values for each dataset
n_values = [
    len(field_scale_precipitation), 
    len(watershed_scale_precipitation), 
    len(field_scale_total_flow), 
    len(watershed_scale_total_flow)
]

# Labels for the boxplot groups (including n= values)
labels = [
    f"Field scale:\n % of Precipitation\n(n={n_values[0]})",
    f"Watershed scale:\n % of Precipitation\n(n={n_values[1]})",
    f"Field scale:\n % of Total Flow\n(n={n_values[2]})",
    f"Watershed scale:\n % of Total Flow\n(n={n_values[3]})"
]

# Colors: One shade of blue for Precipitation, another for Total Flow
colors = ['#87ceeb', '#87ceeb', '#4682b4', '#4682b4']  # Sky blue for precipitation, steel blue for flow

# Dictionary to store statistics
statistics = {}

# Calculate statistics for each dataset
for label, dataset in zip(labels, data):
    dataset = np.array(dataset)
    statistics[label] = {
        "Minimum": np.min(dataset),
        "Lower Quartile (Q1)": np.percentile(dataset, 25),
        "Median": np.median(dataset),
        "Upper Quartile (Q3)": np.percentile(dataset, 75),
        "Maximum": np.max(dataset),
    }

# Create the boxplot
fig, ax = plt.subplots(figsize=(12, 6))
box = ax.boxplot(data, patch_artist=True, labels=labels)

# Apply colors to the boxes
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Add title and axis labels
plt.title("Tile Flow as a Percentage of Annual Precipitation and Total Flow", fontsize=14)
plt.ylabel("Percentage (%)", fontsize=12)
plt.xticks(fontsize=12)  # Ensure no rotation for x-axis tick labels
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Add a legend for color distinction
plt.legend([box['boxes'][0], box['boxes'][2]], 
           ['Precipitation', 'Total Flow'], 
           loc='upper right', fontsize=10, frameon=False)

# Save the plot as a JPG file with 300 dpi
output_path = r"C:\Users\mbodrudd\OneDrive - University of Guelph\Zion\Zion PhD study\PhD articles writing\Tile_Drain_Review\Final_Revised_Files_Jan01\YangReviewed5Jan\Review_Jan01_2025\Need to submit 8 Jan\Figure_3.jpg"
plt.tight_layout()
plt.savefig(output_path, format='jpg', dpi=300)

# Show the plot
plt.show()

# Print the statistics
print("Box Plot Statistics:")
for label, stats in statistics.items():
    print(f"\n{label}:")
    for stat_name, stat_value in stats.items():
        print(f"  {stat_name}: {stat_value:.2f}")
```
